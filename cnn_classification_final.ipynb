```python
from PIL import Image
import os, glob
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
from keras.models import Sequential
from keras.layers import Convolution2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
import numpy as np
```

```python
caltech_dir="./instagram_O/2"
caltech_dirr="./instagram_O/TEST"
categories=["coat","dress","pullover","shirt","t"]
nb_classes=len(categories)
```

```python
image_w=64 
image_h=64
```

### TEST 데이터 불러오기

```python
xx=[]
yy=[]
for idx, cat in enumerate(categories):
    # 레이블 지정 ---4
    label=[0 for i in range(nb_classes)]
    label[idx]=1
    # 이미지 ---5
    image_dirr=caltech_dirr+"/"+cat
    files=glob.glob(image_dirr+"/*.*")
    for i, f in enumerate(files):
        img=Image.open(f) # ---6
        img=img.convert("RGB")
        img=img.resize((image_w, image_h))
        data=np.asarray(img)
        xx.append(data)
        yy.append(label)
        if i%10==0:
            print(i, "\n", data)
xx=np.array(xx)
yy=np.array(yy)
```

### TRAIN 데이터 불러오기

```python
x=[]
y=[]
for idx, cat in enumerate(categories):
    # 레이블 지정 ---4
    label=[0 for i in range(nb_classes)]
    label[idx]=1
    # 이미지 ---5
    image_dir=caltech_dir+"/"+cat
    files=glob.glob(image_dir+"/*.*")
    for i, f in enumerate(files):
        img=Image.open(f) # ---6
        img=img.convert("RGB")
        img=img.resize((image_w, image_h))
        data=np.asarray(img)
        x.append(data)
        y.append(label)
        if i%10==0:
            print(i, "\n", data)
x=np.array(x)
y=np.array(y)
```

### 데이터 셋 분리 및 정규화

```python
# 학습 전용 데이터와 테스트 전용 데이터 구분 ---101
x_train=x
x_test=xx
y_train=y
y_test=yy
xy=(x_train, x_test, y_train, y_test)
np.save("./instagram_O/5obj.npy", xy)

print("ok,", len(y))
```

```python
categories=["coat","dress","pullover","shirt","t"]
nb_classes=len(categories)
image_w=64
image_h=64
```

```python
x_train, x_test, y_train, y_test = np.load("./instagram_O/5obj.npy", allow_pickle=True)
#데이터 정규화하기
x_train = x_train.astype("float")/255
x_test = x_test.astype("float")/255
print('x_train:', x_train.shape)
len(y_train)
```

### 모델 학습(1)

```python
import keras.backend.tensorflow_backend as K
with K.tf.device('/device:GPU:2'):
    model = Sequential()
    model.add(Convolution2D(32,3,3,border_mode='same',input_shape=x_train.shape[1:]))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(0.25))
    model.add(Convolution2D(64, 3, 3, border_mode='same'))
    model.add(Activation('relu'))
    model.add(Convolution2D(64, 3, 3, border_mode='same'))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(0.25))
    model.add(Flatten())
    model.add(Dense(512))
    model.add(Activation('relu'))
    model.add(Dropout(0.5))
    model.add(Dense(nb_classes))
    model.add(Activation('softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    #모델 훈련하기
    model.fit(x_train, y_train, batch_size=50, nb_epoch=10)
```

### total accuracy

```python
score = model.evaluate(x_test,y_test, verbose=0)
print('loss=', score[0])
print('accuracy=',score[1])
```

### 개별 accuracy

```python
categories=["coat","dress","pullover","shirt","t"]
for i in range(5):
    yyy_test=[]
    xxx_test=[]
    for j in range(5000):
        if y_test[j][i]==1:
            yyy_test.append(y_test[j])
            xxx_test.append(x_test[j])
    yyy_test=np.array(yyy_test)
    xxx_test=np.array(xxx_test)
    score = model.evaluate(xxx_test,yyy_test, verbose=0)
    print(categories[i]+'의 loss=', score[0])
    print(categories[i]+'의 accuracy=',score[1])
    print('-----------------------------------')
```
